# Minimum skew problem

```python
def min_skew(text):
    count=0
    min_count=0
    min_loc=[]
    for i in range(len(text)):
        if text[i]=="G":
            count+=1
        elif text[i]=="C":
            count+=-1
        if count<min_count:
            min_loc=[]
            min_loc.append(i)
            min_count=count
        elif count==min_count:
            min_loc.append(i)
    return(min_loc)
```

```python
def pattern_matching(pattern,text,d):
    pos=[]
    for i in range(len(text)-len(pattern)+1):
        kmer=text[i:i+len(pattern)]
        l=hamming_distance(kmer,pattern)
        if l<=d:
            pos.append(i)
    print(len(pos))
    return(pos)
```

# Finds the no. of kmers in genome that differ with a particular kmer by only d bases.

```python
def pattern_matching_no(pattern,text,d):
    pos=[]
    for i in range(len(text)-len(pattern)+1):
        kmer=text[i:i+len(pattern)]
        l=hamming_distance(kmer,pattern)
        if l<=d:
            pos.append(i)
    return(len(pos))
```

# Finds the most frequent kmer in genome ie. usually near and about the oric , probably is the kmer that accounts for DNAa boxes.  

```python
def frequent_kmer(text,k,d):
    max_count=0
    count=0
    freq_kmer=[]
    kmer=""
    for i in range(len(text)-k+1):
        kmer=text[i:i+k]
        count=pattern_matching_no(kmer,text,d)
        if count>max_count:
            freq_kmer=[]
            max_count=count
        elif count==max_count:
            freq_kmer.append(kmer)
    return(freq_kmer)
```

```python
def frequent_kmer_with_rc(text,k,d):
    min_count=0
    count=0
    freq_kmer=[]
    kmer_list=kmer_generator(k)
    for i in kmer_list:
        kmer=i
        count=pattern_matching_no(kmer,text,d)+pattern_matching_no(reverse_complement(kmer),text,d)
        if (count>min_count):
            freq_kmer=[]
            freq_kmer.append(kmer)
            min_count=count
        elif count==min_count:
            freq_kmer.append(kmer)
    return(freq_kmer)
```

```python
def reverse_complement(text):
    rc=""
    for i in text:
        if i=="A":
            rc+="T"
        if i=="C":
            rc+="G"
        if i=="G":
            rc+="C"
        if i=="T":
            rc+="A"
    return(rc[::-1])
```

```python
import itertools as it
def kmer_generator(k):
    x=list(it.product("ATGC",repeat=k))
    return(x)
```

```python
def neighbors(pattern,d):
    k=len(pattern)
    kmer=kmer_generator(k)
    neigh=[]
    for i in kmer:
        if (hamming_distance(i,pattern)<=d):
            neigh.append(i)
    neigh=list(dict.fromkeys(neigh))
    return(neigh)
```
