# Generating Theoretical Spectrum Problem: Generate the theoretical spectrum of a cyclic peptide

```python
def mass(peptide):
    m=0
    for i in peptide:
        m+=y[i]
    return(m)
def linearspectrum(peptide):
    l=[0]
    for i in range(1,len(peptide)+1):
        l.append(mass(peptide[0:i]))
    ls=[0]
    for j in range(len(l)):
        for k in range(1+j,len(l)):
            x=(l[k]-l[j])
            ls=ls+[x]
    ls.sort()
    return(ls)
def cyclospectrum(peptide):
    spectrum = [0]
    for i in range(0,len(peptide)):
        for j in range(1,len(peptide)):
            spectrum.append(mass(peptide[:j]))
        x=peptide[0]
        peptide=peptide[1:]+x
    spectrum.append(mass(peptide))
    spectrum.sort()
    return(spectrum)    
```

```python
def expand(candidatepeptides):
    aminoacids=["G","A","S","P","V","T","C","I","L","N","D","K","Q","E","M","H","F","R","Y","W" ]
    new=[]
    for i in candidatepeptides:
        a=[]
        for j in aminoacids:
            a.append(i+j)
        new.extend(a)
    return(new)
def parentmass(spectrum):
    return(max(spectrum))
def consistent(peptide,spectrum):
    k=1
    for i in cyclospectrum(peptide):
        if i in spectrum:
            spectrum.remove(i)
            k=k*1
        else:
            k=k*0
            break
    if k==1:
        return(True)
    elif k==0:
        return(False)
```

```python
def cyclopeptide_sequencing(spectrum):
    candidatepeptides=[""]
    finalpeptides=[]
    while len(candidatepeptides)>0:
        candidatepeptides=expand(candidatepeptides)
        print(candidatepeptides)
        for peptide in candidatepeptides:
            if mass(peptide)==parentmass(spectrum):
                if (cyclospectrum(peptide)==spectrum) and (peptide not in finalpeptides):
                    finalpeptides.append(peptide)
                candidatepeptides.remove(peptide)
            elif consistent(peptide,spectrum)==False:
                candidatepeptides.remove(peptide)
        print(finalpeptides)
    return(finalpeptides)
```
