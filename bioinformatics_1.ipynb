# #pattern count : counts the no. of times a particular pattern ie. a kmer occurs in a genome.

```python
def patterncount(text,pattern):
    count=0
    for i in range(len(text)-len(pattern)):
        if(text[i:i+len(pattern)]==pattern):
            count=count+1
    return(count)
```

# # frequent words: tells which is the most frequent kmer in a piece of genome, if we find the oric the most frequent kmer is going to probably be DNAa box.

```python
def frequentwords(text,k):
    count=0
    maxcount=0
    frequentpatterns=[]
    for i in range(0,len(text)-k):
        pattern=text[i:i+k]
        count=patterncount(text,pattern)
        if(count==maxcount):
            if(pattern not in frequentpatterns):
                frequentpatterns.append(pattern)
        if(count>maxcount):
            if(pattern not in frequentpatterns):
                frequentpatterns=[]
                frequentpatterns.append(pattern)
                maxcount=count
    return(frequentpatterns)           
            
```

## reverse complement: finds the complement of a dna strand and gives its complement

```python
def reverse_complement(text):
    rc=""
    for i in text:
        if i=="A":
            rc+="T"
        if i=="C":
            rc+="G"
        if i=="G":
            rc+="C"
        if i=="T":
            rc+="A"
    return(rc[::-1])
```

## computing frequencies: by using hashmap or dictionary data type 

```python
def pattern_generator(pattern):
    value=0
    K=len(pattern)
    for i in range(len(pattern)):
        if pattern[i]=="A":
            value+=0
        elif pattern[i]=="T":
            value+=(4^(K-i-1))*3
        elif pattern[i]=="G":
            value+=(4^(K-i-1))*2
        else:
            value+=(4^(K-i-1))*1
    return(value)    
```
